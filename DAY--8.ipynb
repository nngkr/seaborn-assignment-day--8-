LINE PLOT

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data={
'x':['a','b','c','d'],
'y':[12,20,18,23]
}
df=pd.DataFrame(data)
sns.lineplot(x='x',y='y',data=df,
             color='red',
             linestyle='--',
             marker='o')
plt.show()
```

BAR PLOT

```python
data={
    'student':['krishna','shiva','vikram','eswar'],
    'maths':[21,15,19,24],
    'english':[17,19,21,16]
}
df=pd.DataFrame(data)
print(df)
sns.set_style('darkgrid')
sns.barplot(x='student',y='maths',data=df,
            color='red',
            edgecolor='black',
            )
plt.title("STUDENT MARKS")
plt.show()
```

HISTOGRAM PLOT

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data={
    'name':['krishna','supriya','pranathi','madhu','vishnu'],
    'attendance':[65,93,81,85,95]
    
}
df=pd.DataFrame(data)
print(df)
sns.set_style('ticks')
sns.histplot(x='name',y='attendance',data=df,bins=5,
             color='green',
             edgecolor='black')
plt.title("ATTENDANCE OF THE STUDENTS")
plt.show()
```

SCATTER PLOT

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data={
    'hours':[1,3,2,6,2],
    'marks':[16,23,18,29,27],
    'gender':['male','female','male','female','male']
}
df=pd.DataFrame(data)
sns.set_style('whitegrid')
sns.scatterplot(x='hours',y='marks',data=df,
                hue='gender',
                linestyle='--',
                marker='*',
                style="gender")
plt.title("HARD WORK")
plt.show()
```

VIOLIN PLOT

```python
data={
    'name':['a','b','c','d','e','f','g','h','i','j'],
    'marks':[32,31,30,40,41,42,43,44,45,46]
}
df=pd.DataFrame(data)
sns.violinplot(x='marks',data=df,
               color='blue',
               inner='box')
plt.title("RANGE OF THE STUDENT MARKS")
plt.show()
```

SWARM PLOT

```python
data ={'name':['sunshma','sunitha','radhika','vikram','pavani','apurupa'],
       'Age':[20,24,25,26,25,27]
       }
df=pd.DataFrame(data)
print(df)
sns.swarmplot(x='name',y='Age',data=df,
              size=10,
              hue='name')
plt.title('AGE AND NAMES')
plt.show()
```

BOX PLOT

```python
data = {
    'Student': [
        'Ram', 'Sita', 'Ravi', 'Hari', 'Kiran',
        'Asha', 'Venu', 'Radha', 'Mohan', 'Teja'],
    

    'Subject': [
        'Math', 'Math', 'Math', 'Math', 'Math',
        'Science', 'Science', 'Science', 'Science', 'Science'],
    

    'Marks': [
        45, 50, 55, 60, 95,
        40, 42, 45, 48, 90]
    
}
df = pd.DataFrame(data)
print(df)
plt.figure(figsize=(8,5))
sns.boxplot(
    x='Subject',
    y='Marks',
    data=df,
    hue='Subject'
    


)
plt.title("Student Marks Analysis Using Box Plot")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```
